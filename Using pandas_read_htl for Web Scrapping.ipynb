```python
year = "2021"
url_link = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"

# Combing the the url
url = url_link.format(year)
url
```

```python
# importing pandas library
import pandas as pd
```

```python
# using pd.read_html() function to read NTML 
basktbll = pd.read_html(url, header = 0)
basktbll[0]
```

```python
bb2022 = basktbll[0]
bb2022
```

### Deleting multiple headers 

```python
bb2022[bb2022.Age =="Age"]
```

```python
#Number of rows with the headers
len(bb2022[bb2022.Age == 'Age'])
```

```python
bb2022.shape
```

```python
bb2022=bb2022.drop(bb2022[bb2022.Age == 'Age'].index)
bb2022.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style("darkgrid")
sns.color_palette("deep")
sns.color_palette("pastel")
sns.color_palette("Set2")
plt.rc("axes" , titlesize=17) # fontsize of the axis title
plt.rc("axes", labelsize=13)   #fontsize of x and y labels
plt.rc('xtick', labelsize=12)  #fontsize of the tick labels
plt.rc('ytick', labelsize=12)  #font of the tick labels
plt.rc("legend", fontsize=12)  #legend font size
plt.rc('font', size=12)  
bins = ( 19, 21, 23,  25, 27, 29, 31, 33, 35, 37,39)
plt.figure(figsize=(10, 7), tight_layout = True)
ax = sns.distplot(bb2022.Age, bins = bins, kde  = False,color= "blue")
ax.set( title = "Age Histogram", xlabel = "Age", ylabel = "Count")
```
